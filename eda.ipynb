# Finding the Optimal Number of Clusters

```python
# find the optimal number of clusters using the Elbow method
from sklearn.cluster import KMeans
import numpy as np
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(rfm_df)
    inertia.append(kmeans.inertia_)
plt.figure(figsize=(12, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.grid()
plt.show()
```

```python
# let's run kmeans with n_clusters = 3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(rfm_df)
df['cluster'] = kmeans.labels_
# Visualize the clusters
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='number_of_days_since_last_purchase', y='total_amount_spent', hue='cluster', palette='viridis')
plt.title('Clusters of Customers Based on RFM Features')
plt.xlabel('Number of Days Since Last Purchase')
plt.ylabel('Total Amount Spent')
plt.show()
```

### Cluster 1: Low-Value Customers
This group is made up of low-spending, infrequent customers.

 - Recency: They are moderately recent.

 - Monetary: This is their defining feature. They spend very little (average spend <100).

 - Frequency: They purchase very infrequently.

 - Response Rate: They have the lowest response rate (~9%), making them unattractive targets for campaigns.

### Cluster 2: Best Customers (Champions)
This is your most valuable segment. They are recent, frequent, and high-spending customers who are highly engaged.

 - Recency: This is their defining feature. They have purchased very recently (average <25 days).

 - Monetary: They are high spenders.

 - Frequency: They buy frequently.

 - Response Rate: They have an exceptionally high response rate (~28%), making them the perfect target for new offers.

# Phase 3: Predictive Modeling
The final key objective is to "Create a predictive model which allows the company to maximize the profit of the next marketing campaign". 

The goal is to teach a machine learning model to recognize the patterns of a "responder" so we can "cherry pick" the best customers to contact in the next campaign.

```python
#logistic regression
# define the features of x
x = df[['age', 'income', 'kids_at_home', 'teens_at_home', 'amount_spent_on_fish_products',
        'amount_spent_on_fruits', 'amount_spent_on_gold_products', 'amount_spent_on_meat_products',
        'amount_spent_on_regular_products', 'amount_spent_on_sweet_products', 'amount_spent_on_wines',
        'number_of_purchases_made_using_catalogs', 'number_of_purchases_made_using_deals',
        'number_of_purchases_made_in_store', 'number_of_purchases_made_through_website',
        'number_of_website_visits_last_month', 'number_of_days_since_last_purchase',
        'basic_level_education', 'graduation_level_education', 'masters_level_education',
        'phd_level_education', 'second_cycle_level_education', 'divorced_marital_status',
        'married_marital_status', 'single_marital_status', 'together_marital_status',
        'widowed_marital_status']]
# define the target variable y
y = df['response']
```

```python
# divide the dataset into training and testing sets; 80% for training and 20% for testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(x_train, y_train)
# make predictions on the test set
y_pred = logistic_model.predict(x_test)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# calculate accuracy
# Technical Evaluation: How accurate is the model in its predictions?
# Business Evaluation: How much profit can this model generate?

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
# print classification report
print(classification_report(y_test, y_pred))
# print confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Responded', 'Responded'], yticklabels=['Not Responded', 'Responded'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Understanding Your Classification Report
The report translates the matrix into percentages:

Accuracy (87%): This number looks high, but it's misleading. Your model gets a high score simply because most people (around 85%) are non-responders anyway. It's good at predicting the majority.

Precision (0.61 for Responders): This tells us that when your model predicts someone will respond, it's only right 61% of the time. The other 39% are costly "False Positives."

Recall (0.30 for Responders): This is the most critical metric for our business problem. It means your model only found 30% of all the potential customers in the test set. We missed out on the other 70% (the 44 people in the False Negatives box).

The Business Summary: Your current model is very conservative. It's good at identifying definite non-responders, but it's too cautious and misses out on a huge number of willing customers.

The Final Step: Calculating Profitability
Now for the ultimate test, which directly addresses the key objective to "maximize the profit". Let's calculate the profit your model would have generated on this test set.


From the case study, we know:


Cost per contact: 3 MU (6,720 MU / 2,240 customers) 


Revenue per sale: ~11 MU (3,674 MU / (2,240 * 15% success)) 

Using your confusion matrix:

Customers you would contact: True Positives + False Positives = 19 + 12 = 31 customers

Total Cost: 31 customers * 3 MU/customer = 93 MU

Total Revenue: 19 successful sales * 11 MU/sale = 209 MU

```python
#let's redo the logistic regression model
x = df[['age', 'income', 'kids_at_home', 'teens_at_home',
        'number_of_days_since_last_purchase', 'frequency', 'total_amount_spent']]

y = df['response']
```

```python
# divide the dataset into training and testing sets; 80% for training and 20% for testing
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(x_train, y_train)
# make predictions on the test set
y_pred = logistic_model.predict(x_test)
# calculate accuracy
# Technical Evaluation: How accurate is the model in its predictions?
# Business Evaluation: How much profit can this model generate?

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
# print classification report
print(classification_report(y_test, y_pred))
# print confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Responded', 'Responded'], yticklabels=['Not Responded', 'Responded'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
